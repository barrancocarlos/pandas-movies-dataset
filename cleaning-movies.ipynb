```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 20, 'figure.figsize': (10, 8)})
```

```python
#Import data file
data = pd.read_csv("cleaning-movies.csv", index_col="Film") 
```

```python
#Convert to dataframe
main_table = pd.DataFrame(data)
```

```python
#Get basic info
main_table.info()
```

```python
#Display dataset
main_table.head(20)
```

```python
#Display Profitability 
main_table['Profitability']
```

```python
#Check Profitability for types of data
main_table['Profitability'].apply(type)
```

```python
#Search for entry that prevents from converting series to floats
main_table.loc[main_table['Profitability'] == 'six thousand']
```

```python
#Replace value to be able to convert to float
main_table.loc["What Happens in Vegas"] = main_table.loc["What Happens in Vegas"].replace('six thousand', "6000") 
```

```python
# Remove commas and convert string to float
main_table['Profitability'] = main_table['Profitability'].str.replace(',', '').astype(float)
```

```python
#Check Profitability for types of data again
main_table['Profitability'].apply(type)
```
